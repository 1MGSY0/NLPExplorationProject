<h4>Summarisation with Textrank</h4>

```python
test_df = pd.read_csv("C:/Users/tanwe/Desktop/School Stuff/GitHub/NLPExplorationProject/data/Dataset1/test_summary.csv", sep="\t")
```

```python
import spacy
import pytextrank

def summarise_reviews(df, col, lim_phrase, lim_sent): 
    nlp = spacy.load("en_core_web_lg")
    nlp.add_pipe("textrank")   

    summarised = []
    sentences = 0
    index = 0
    for review in df[col]:
        doc = nlp(review)

        for sent in doc._.textrank.summary(limit_phrases = lim_phrase, limit_sentences = lim_sent):
            if sentences == 0:
                summarised += [sent.text]
            else:
                summarised[index] += sent.text
            sentences += 1

            if sentences == lim_sent:
                index += 1
                sentences = 0
    return summarised
            
summarised_lst = summarise_reviews(test_df, "review/text", 5, 5)  
print(len(summarised_lst))
```
