<h3>Summarisation with Textrank</h3>

```python
# Import packages
import pandas as pd
import spacy
import pytextrank
import evaluate
```

```python
df = pd.read_csv("C:/Users/tanwe/Desktop/School Stuff/GitHub/NLPExplorationProject/data/Dataset1/test_summary.csv", sep="\t")
```

```python
def summarise_reviews(df, col, lim_phrase, lim_sent):
    nlp = spacy.load("en_core_web_lg")
    nlp.add_pipe("textrank")

    summaries = []
    for review in df[col]:
        doc = nlp(review)
        summary = " ".join(sent.text 
            for sent in doc._.textrank.summary
            (
                limit_phrases=lim_phrase, limit_sentences=lim_sent
            )
        )
        summaries.append(summary)

    return summaries

pred_summarise = summarise_reviews(df, "review/text", 5, 5)  
```

```python
print(pred_summarise[1])
```

```python
ref_summary = []
for summary in df['review/summary']:
    ref_summary.append(summary)
```

```python
rouge = evaluate.load("rouge")
results = rouge.compute(predictions = pred_summarise, references = ref_summary, use_aggregator=True)
results = {k: float(v) for k, v in results.items()}
print(results)
```
